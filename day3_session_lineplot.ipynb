```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset in nupyter notebook

df = sns.load_dataset("iris")
df
```

```python
# Line plot and bar plot for iris datset 
sns.lineplot(x="sepal_length", y="sepal_width", data= df)
plt.show()
```

```python
# we can limit the x and y 
sns.lineplot(x="sepal_length", y="sepal_width", data= df)
plt.xlim(2)
plt.show()

```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data= df)
plt.title(" This is line plot and bar plot")
plt.show()
```

```python
# next ploting the different plots function 
sns.lineplot(x="sepal_length", y="sepal_width", data= df)
import seaborn as sns
import matplotlib.pyplot as plt
plt.title(" This is line plot and bar plot")
plt.figure(figsize=(4,2))
sns.barplot(x="species", y="sepal_width", data= df)
plt.show()


```

```python
# Import
df_raw = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")

# Prepare Data
df = df_raw.groupby('class').size()

# Make the plot with pandas
df.plot(kind='pie', subplots=True, figsize=(8, 8) )
plt.title("Pie Chart of Vehicle Class - Bad")
plt.ylabel("")
plt.show()
```

```python


import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```
